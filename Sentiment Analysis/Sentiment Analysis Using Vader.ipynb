```python
from nltk import *
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
```

```python
# cleaning, tokenizing, removing stop words, lemmatization and returning frequency distribution 
def clean_review(li):
    a=str(li.tolist()).lower()
    txt=re.sub(r'[^a-z]+',' ',str(a)).strip()
    tokens=word_tokenize(txt)
    stop_words=stopwords.words('english')
    words=[t for t in tokens if t not in stop_words]
    lem=WordNetLemmatizer()
    l=[lem.lemmatize(w) for w in words]
    fd=FreqDist(l)
    return fd
```

```python
fd=clean_review(df['Review_text'])
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sent=SentimentIntensityAnalyzer()
```

```python
polarity=[sent.polarity_scores(i)['compound'] for i in df['Review_text']]
```

```python
# adding polarity scores to df
df['polarity']=polarity
```

```python
df
```

```python
def tags(x):
    if x<=-0.05:
        return 'negative'
    elif -0.05<x<0.05:
        return 'neutral'
    else:
        return 'positive'
```

```python
# based on polarity score, tagging as positive or negative
df['tags']=df['polarity'].map(tags)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.feature_extraction.text import TfidfVectorizer, TfidfTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
train,test=train_test_split(df["Review_text"],test_size=0.2,random_state=1)
```

```python
train=[re.sub(r'[^a-zA-z0-9]',' ',i) for i in train]
test=[re.sub(r'[^a-zA-z0-9]',' ',i) for i in test]
```

```python
# tokenizing and LEmatizing by defining a class
class LemmaTokenizer(object):
    def __init__(self):
        self.wnl=WordNetLemmatizer()
    def __call__(self,doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc)]
```

```python
tfidf=TfidfVectorizer(max_features=10000,stop_words="english",ngram_range=(1,1),tokenizer=LemmaTokenizer(),max_df=0.8)
tr_tfidf=tfidf.fit_transform(train)
```
